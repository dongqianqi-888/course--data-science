# pivot table and cross tabulation


```python
data={
    'city':['kl','kunming','kl','kunming','kl'],
    'category':['a','a','b','b','a'],
    'sales':[2323,5324,6453,6248,6456]}
df=pd.DataFrame(data)
df
```

```python
pd.pivot_table(df,values='sales',index='city',columns='category')
```

```python
pd.pivot_table(df,values='sales',index=['city','category'])
```

```python
pd.pivot_table(df,values='sales',index='city',aggfunc='sum')
```

```python
pd.crosstab(index=df['city'],columns=df['category'])
```

# view and copy 

```python
df1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]},index=['a','b','c'])
view=df1['A']
view[0]=9999
print(df1)
print(view)
```

```python
df1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]},index=['a','b','c'])
view=df1['A'].copy()
view[0]=9999
print(df1)
print(view)
```

```python
df1=pd.DataFrame({'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]},index=['a','b','c'])
subset=df1[df1['A']>1].copy()
subset['B']=9999
print(df1)
print(subset)
```
