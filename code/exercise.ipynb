```python
iris=sns.load_dataset('iris')
iris.loc[:,'size_category']=pd.cut(iris.loc[:,'petal_length'],bins=[0,2.5,5,max(iris.loc[:,'petal_length'])],labels=['small','medium','large'])
iris.head(5)
```

```python
iris.iloc[:,0:5].groupby('species').agg(['mean','min','max'])
```

```python
iris.loc[:,'sepal_ratio']=iris.loc[:,'sepal_length']/iris.loc[:,'sepal_width']
iris.loc[:,'petal_ratio']=iris.loc[:,'petal_length']/iris.loc[:,'petal_width']
iris.iloc[:,-2:].corr()
```

```python
pd.pivot_table(iris,index='species',columns='size_category',values='petal_length',aggfunc='count')
```

```python
import numpy as np
import pandas as pd
data = np.array([
    ["Alice", 25, 90],
    ["Bob", 35, 85],
    ["Charlie", 28, 95],
    ["David", 45, 80],
    ["Eve", 22, 92]
])
df = pd.DataFrame(
    data,
    columns=['name', 'age', 'score'],  # 列名（3个，与数据列数匹配）
    index=np.arange(0, 5, 1)  # 索引0-4（5个，与数据行数匹配，修正你的4→5）
)
df['age']=df['age'].astype(int)
df['score']=df['score'].astype(int)

def dis(x):
    if x > 50:
        return '老年'
    elif x < 30:
        return '青年'
    else:
        return '中年'


# 3. map函数应用+深拷贝（修正：函数调用，统一引号）
age_new = df['age'].map(dis).copy()  # 正确，保留

age_new
```
