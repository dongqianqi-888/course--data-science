# duplicates

```python
#duplicates
data={
    'name':['donna','jenny','zahra','mica','donna'],
    'age':['22','23','24','21','22'],
    'city':['kl','jb','la','kl','kl']
}
df=pd.DataFrame(data)
print(df.duplicated())
print(df.drop_duplicates())
```

```python
#subset duplicates
data={
    'name':['donna','jenny','zahra','mica','donna'],
    'age':['22','23','24','21','22'],
    'city':['kl','jb','la','kl','kl2']
}
df=pd.DataFrame(data)
print(df.duplicated(subset='name',keep='first'))
drop=df.duplicated(subset='name',keep='first')
print(df[~drop])
```

# missing

```python
data={
    'a':[1,2,None,4],
    'b':[np.nan,2,3,4],
    'c':['na','valid',None,'missing']
}
df=pd.DataFrame(data)
print(df.isna())
print(df.isnull())
print(df.notna())
```

```python
df['c']=df['c'].replace('na',None)
df
```

```python
#delet all of columns which contains missing value
df1=df.dropna(axis=1)
df1

#delet all of columns which contains missing value
df2=df.dropna(axis=0)
df2

#delet rows with missing value in specific columns
df3=df.dropna(subset=['a'],axis=0)
df3
```

```python
#fill with a specific value
df1=df.fillna(0)
print(df1)

#fill with the previous value
df2=df.ffill()
print(df2)

#fill with the next value
df3=df.bfill()
print(df3)

#fill by linear model
#df4=df.interpolation(method='linear')
#df4
```

# apply&map

```python
data={
    'name':['donna','mica','zahra'],
    'weight':[56,45,65],
    'height':[1.67,1.57,1.77]
}
df=pd.DataFrame(data)
df
```

```python
def bmi(r):
    bmi=r['weight']/r['height']**2
    return bmi
df['bmi']=df.apply(bmi,axis=1)
print(df)
```

```python
df.iloc[:,1:4].apply(sum,axis=1)
```

```python
df=pd.DataFrame({
    'fruit':['apple','pear','banana']
})
cate={'apple':'imported','pear':'imported','banana':'local'}
df['type']=df['fruit'].map(cate)
df
```

# outliers

```python
df=pd.DataFrame({'value':[3343,3256,3643,999,3322,3345,3224,3451,3641]})
print(df)

mean=df['value'].mean()
std=df['value'].std()
upper=mean+3*std
lower=mean-3*std

out=(df['value']>upper)|(df['value']<lower)
print(out)
```

```python
df['value']=np.where(out,999,df['value'])
df
```

## discretization

```python
#equal ranges
data={'mark':[54,35,34,24,13,51,54]}
df=pd.DataFrame(data)
df
```

```python
df['disc']=pd.cut(df['mark'],bins=3)
df
```

```python
df['disc']=pd.cut(df['mark'],bins=[0,20,40,60,80,100])
df
```

```python
df['cate']=pd.cut(df['mark'],bins=[0,20,40,60,80,100],labels=[1,2,3,4,5])
df
```

```python
#equal frequencies
df['discq']=pd.qcut(df['mark'],q=4)
df
```

## permutation&sampling

```python
arr=np.array([1,2,3,5,3,6,8,5])
np.random.seed(123)
per_arr=np.random.permutation(arr)
print(arr)
print(per_arr)
```

```python
df=pd.DataFrame({
    'A':[1,2,3,4,5],
    'B':['a','s','d','f','g']})
per_df=df.sample(frac=1,random_state=123)
print(df)
print(per_df)
```

```python
#sampling
arr=np.array([1,2,3,5,3,6,8,5])
np.random.seed(123)
se_arr=np.random.choice(arr,size=3,replace=False)
print(arr)
print(se_arr)
```

```python
df=pd.DataFrame({
    'A':[1,2,3,4,5],
    'B':['a','s','d','f','g']})
se_df=df.sample(frac=0.4,random_state=123)
se_df2=df.sample(n=4,replace=True,random_state=123)
print(df)
print(se_df)
print(se_df2)
```
