# for loop

```python
dic={'name': 'donna', 'age': '23', 'city': 'lanzhou'}
for key, value in dic.items():
    print('\nkey:', key) 
    print('value:', value) 
```

```python
for value in dic.values():
    print(value)
```

```python
a1={'time':'2 nov','event':'dm video'}
a2={'time':'5 nov','event':'dm pre'}
a3={'time':'6 nov','event':'ba video'}
a=[a1,a2,a3]
for a in a:
    print(a)
```

```python
lesson={'ba':['thu','morning'],
        'ds':['tue','morning'],
        'sta':['wed','morning'],
        'dm':['wed','morning']}
for key,value in lesson.items():
    print(key)
    for info in value:
        print(info)
```

```python
lesson={'ba':{'time':'thu'},
        'ds':{'time':'tue'},
        'sta':{'time':'wed'},
        'dm':{'time':'wed'}}
for key,value in lesson.items():
    print('\nname is '+key)
    print('we need to go on '+value['time'])
```

# while loop 

```python
age=10
while age<=15:
    print('you are '+str(age)+' years old')
    age+=1
```

```python
date=1
while date>0:
    print('today is '+str(date)+'th, hurry up!')
    date+=1
    if date>5:
        break
```

```python
day=7
while day>0:
    print(str(day)+' day(s) until the holiday!')
    day-=1
    if day==0:
        print('today is a holiday!')
```

```python
while True:
    info=input('what is your name?')
    if info==' ':
        break
    else:
        print(info+' is a good guy. we all like '+info)
```

```python
active=True
while active:
    n=input('what do you do in your free time?')
    print('wow, but '+n+' runs countercurrent to my natrue.')
    if n==' ':
        active=False
    
```
