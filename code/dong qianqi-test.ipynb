## 1. accept number from users and display the grade.

```python
m=input('Please input your marks.')
m=float(m)
if m>90: 
    print('You get A!')
elif 80<=m:
    print('You get B!')
elif 60<=m:
    print('You get C.')
else:
    print('You get D~')
```

## 2. nutrition analyst   

a) dictionary 

```python
dietary={'Donna':'roti_canel','Zahra':'nasi_lemak','Mica':'nasi_goring','Cici':'mee_goring','Jenny':'beef_rice'}
print(dietary)
```

b) use loop to describe 

```python
for name,food in dietary.items():
    print(name+"'s favorite food is "+food+'.')
```

c) contain all the food into list



```python
favorite_foods=[]
for value in dietary.values():
    favorite_foods.append(value)
```

## 3. the secret number 

```python
num=45
while True:
    user=int(input('Please guess the secret number which between 10 and 50!\n'))
    if user<10|user>50:
        print('You input the wrong number, please try again!\n')

    elif user==num:
        print('Well down! You get it!\n')
        break
    else:
        print('Incorrect, try again!\n')
        continue
```
