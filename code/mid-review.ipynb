```python
#class
class my:
    def __init__(self,energy,clean,mood):
        self.energy=energy
        self.clean=clean
        self.mood=mood
        self.total=0
    def des(self):
        print('the energy score is: '+str(self.energy))
        print('the clean score is: '+str(self.clean))
        print('the mood score is: '+str(self.mood))
    def sleep(self):
        self.energy+=10
        self.clean-=3
        self.mood+=8
        self.total=self.energy+self.clean+self.mood+self.total
        print('the total score is: '+str(self.total))
    def exercise(self):
        self.energy+=8
        self.clean-=5
        self.mood+=8
        self.total=self.energy+self.clean+self.mood+self.total
        print('the total score is: '+str(self.total))
    def shower(self):
        self.energy+=3
        self.clean+=20
        self.mood+=5
        self.total=self.energy+self.clean+self.mood
        print('the total score is: '+str(self.total))
    def eat(self):
        self.energy+=10
        self.clean-=5
        self.mood+=10
        self.total=self.energy+self.clean+self.mood
        print('the total score is: '+str(self.total))
```

```python
m=my(0,0,0)
m.sleep()
m.shower()
m.eat()
```

```python
m.des()
```

```python
class my_dietary(my):
    def __init__(self,energy,clean,mood):
        super().__init__(energy,clean,mood)
        self.storing=0
    def buy(self):
        self.storing+=10
        self.energy-=5
        self.clean-=5
        self.mood+=8
        self.total=self.energy+self.clean+self.mood+self.storing
        print('the total score is: '+str(self.total))
    
```

```python
m2=my_dietary(23,12,23)
m2.buy()
m2.des()
```
