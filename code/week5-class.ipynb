```python
class person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def myfun(self):
        print('hello, my name is '+self.name+'~')
```

```python
p1=person('donna',22)
p2=person('mica',25)
```

```python
print(p1.name)
print(p1.age)
```

```python
p2.myfun()
```

```python
class car():
    def __init__(self,make,model,year):
        self.make=make
        self.model=model
        self.year=year
        self.odometer_reading=0
    def descriptive(self):
        long=str(self.year)+' '+self.make+' '+self.model
        return long.title()
    def read_odometer(self):
        print('this car has '+str(self.odometer_reading)+' miles on it.')
    def update_odometer(self,mile):
        self.odometer_reading=mile
    def increment_odometer(self,miles):
        self.odometer_reading+=miles
```

```python
c1=car('audi','a4',2016)
print(c1.descriptive())
```

```python
c1.read_odometer()
```

```python
c1.odometer_reading=23
c1.read_odometer()
```

```python
c1.update_odometer(44)
c1.read_odometer()
```

```python
c1.increment_odometer(20)
c1.read_odometer()
```

```python
class electriccar(car):
    def __init__(self,make,model,year):
        super().__init__(make,model,year)
        self.battery_size=70
    def describe_battery(self):
        print('this car has a '+str(self.battery_size)+'-kWh battery.')
```

```python
c2=electriccar('tesla','model s',2016)
print(c2.descriptive())
c2.describe_battery()
```
