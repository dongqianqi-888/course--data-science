# hierarchical index

```python
df=pd.DataFrame({
    'group':['G1','G2','G1','G2'],
    'subgroup':['a','b','c','d'],
    'value1':[1,2,3,4],
    'value2':[5,6,7,8]})
df
```

```python
df.set_index(['group'])
```

```python
df=df.set_index(['group','subgroup'])
print(df.loc['G1'])
```

```python
#from tuples
index=pd.MultiIndex.from_tuples([
    ('G1','a'),
    ('G1','b'),
    ('G2','a'),
    ('G2','b'),],names=['group','subgroup'])
```

```python
np.random.seed(2323)
data=pd.DataFrame(np.random.randn(4,2),index=index,columns=['v1','v2'])
data
```

```python
#from array
l1=['g1','g1','g2','g2']
l2=['a','b','c','d']

index=pd.MultiIndex.from_arrays([l1,l2],names=['group','subgroup'])
np.random.seed(2323)
data=pd.DataFrame(np.random.randn(4,2),index=index,columns=['v1','v2'])
data
```

```python
#remove index
df1=df.reset_index(level='group')
df1.loc['a']
```

```python
df=pd.DataFrame({
    'group':['G1','G2','G1','G2'],
    'subgroup':['a','b','a','b'],
    'value1':[1,2,3,4],
    'value2':[5,6,7,8]})
gdf=df.groupby(['group','subgroup']).agg({'value1':'sum'})
print(gdf)
```

# long and vide data format

```python
data={
    'student':['ali','siti'],
    'math':[80,78],
    'science':[75,82],
    'english':[89,83]}
data_wide=pd.DataFrame(data)
print(data_wide)
```

```python
#wide to long
df=pd.melt(data_wide,id_vars=['student'],value_vars=['math','science','english'],var_name='subject',value_name='score')
```

```python
#wide to long
data_wide.set_index(['student']).stack()
```

```python
#long to wide
df.set_index(['student','subject']).unstack()
```

```python
#long to wide
df.pivot(index='student',columns='subject',values='score')
```

# merging

```python
df1=pd.DataFrame({'key1':['a','b','c'],'value':[1,2,3]})
df2=pd.DataFrame({'key2':['a','b','c','d'],'value':[4,5,6,7]})
pd.merge(df1,df2,left_on='key1',right_on='key2',how='outer').drop('key2',axis=1)
```

```python
df1=pd.DataFrame({'key1':['a','b','c'],'value':[1,2,3]})
df2=pd.DataFrame({'key2':['a','b','c','d'],'value':[4,5,6,7]})
df2=df2.rename(columns={'key2':'key1'})
pd.merge(df1,df2,on='key1',how='outer')
```

# concatenating

```python
df1=pd.DataFrame({'a':[1,2],'b':[3,4]})
df2=pd.DataFrame({'a':[5,6],'b':[7,8]})
pd.concat([df1,df2],axis=1,join='outer')
```
