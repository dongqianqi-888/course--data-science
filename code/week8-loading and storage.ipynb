# XML

```python
from lxml import etree
```

```python
#top level element that contains everything
school=etree.Element('school')

#create a department
de1=etree.SubElement(school,'department')
de1.set('name','geology')

#create a student
st1=etree.SubElement(de1,'student')
st1.set('id','a0001')
st1.text='ali'
a1=etree.SubElement(st1,'age')
a1.text='21'

st2=etree.SubElement(de1,'student')
st2.set('id','a0002')
st2.text='abu'
a2=etree.SubElement(st2,'age')
a2.text='22'

#create a department
de2=etree.SubElement(school,'department')
de2.set('name','mathmatics')

#create a student
st3=etree.SubElement(de2,'student')
st3.text='alice'
a3=etree.SubElement(st3,'age')
a3.text='23'

st4=etree.SubElement(de2,'student')
st4.text='ahmad'
a4=etree.SubElement(st4,'age')
a4.text='24'
```

```python
stree=etree.ElementTree(school)
stree.write('schooltree.xml',pretty_print=True)
```

```python
rtree=etree.parse('schooltree.xml')
xmlprint=etree.tostring(rtree,pretty_print=True,encoding='unicode')
print(xmlprint)
```

```python
#explore xml file without opening
root=rtree.getroot()
print(root.tag)

ch1=root[0]
print(ch1.tag)

gch1=ch1[0]
print(gch1.tag)

ch2=root[1]
print(ch2.tag)
gch2=ch2[0]
print(gch2.text)
```

```python
for ele in root.iter():
    print(ele.tag)
```

# HDF5

```python
import h5py as h5

with h5.File('example.h5','w') as file:
    group=file.create_group('group1')
    group.attrs['description']='this file contains two data save as dataset1 and dataset2'
    group.create_dataset('dataset1',data=np.arange(10))
    group.create_dataset('dataset2',data=np.random.random((3,3)))
```

```python
with h5.File('example.h5','r') as file:
    group=file['group1']
    note=group.attrs['description']
    data1=group['dataset1'][:]
    data2=group['dataset2'][:]
    print(data1)
    print(data2)
    print(note)
```

```python
with h5.File('example.h5','r') as f:
    print(list(f.keys()))
```
