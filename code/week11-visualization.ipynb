```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# fig and ax

```python
x=np.arange(30)
y=x+3*np.random.default_rng(134).standard_normal(30)
fig,ax =plt.subplots()
ax.scatter(x,y)
ax.set(title='scatter plot',xlabel='x',ylabel='y')
fig.show()
```

```python
rng=np.random.default_rng(12345)
data1=rng.normal(0,1,100)
data2=rng.normal(1,1.5,100)
x=np.arange(30)
y=x+rng.normal(0,3,30)
fig,ax=plt.subplot_mosaic([['a','b'],['c','d']])
ax['a'].hist(data1,bins=20)
ax['a'].set_title('histogram')
ax['b'].scatter(x,y)
ax['b'].set_title('scatter plot')
ax['c'].plot(x,y)
ax['c'].set_title('line plot')
ax['d'].boxplot([data1,data2],tick_labels=['group a','group b'])
ax['d'].set_title('box plot')
#plt.savefig('path')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.patches as patches
fig,ax=plt.subplots()
rect=plt.Rectangle((0.2,0.8),0.2,0.1,color='black',alpha=0.3)
circ=plt.Circle((0.7,0.2),0.15,color='blue',alpha=0.7)
pgon=plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color='green',alpha=0.5)
sqr=plt.Rectangle((0.3,0.1),0.2,0.2,color='black',alpha=0.4)
ellipse=patches.Ellipse((0.5,0.5),0.2,0.1)
hexagon=patches.RegularPolygon((0.7,0.8),6,radius=0.1)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
ax.add_patch(sqr)
ax.add_patch(ellipse)
ax.add_patch(hexagon)
```

```python
# Example macro dataset
macro = pd.read_csv('https://bit.ly/3FYxCYZ')
macro
```

```python
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
data
```

```python
log_data=np.log(data).diff().dropna()
log_data
```

```python
fig,ax=plt.subplots()
sns.regplot(x='m1',y='unemp',data=log_data)
r=log_data['m1'].corr(log_data['unemp'])
ax.text(-0.02,0.2,f'R={r:.2f}')
ax.set_title('regression of m1 and unemp')
```

```python
sns.pairplot(log_data,diag_kind='kde',kind='reg',
             plot_kws={'scatter_kws':{'alpha':0.3}})
```

```python
sns.catplot(x='day',y='tip_pct',data=tips[tips.tip_pct<1],hue='time',col='smoker',
            kind='bar',order=['Thur', 'Fri', 'Sat', 'Sun'])
```

```python
sns.catplot(x='day',y='tip_pct',data=tips[tips.tip_pct<1],col='smoker',row='time',
            kind='bar',order=['Thur', 'Fri', 'Sat', 'Sun'])
```
