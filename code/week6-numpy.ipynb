# array_manipulation

```python
arr=np.arange(12)
print(arr)

# reshape
re_arr=np.reshape(arr,(3,4))
print(re_arr)
re_shape=np.reshape(arr,(2,2,3))
print(re_shape)
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.concatenate((arr1,arr2))
print(arr)
```

```python
arr3=np.array([[1,2],[5,6]])
arr4=np.array([[4,5],[7,8]])
con=np.concatenate((arr3,arr4),axis=1)
print(con)
con=np.concatenate((arr3,arr4),axis=0)
print(con)
```

```python
con=np.vstack((arr3,arr4))
print(con)
```

```python
con=np.hstack((arr3,arr4))
print(con)
```

```python
arr=np.arange(10)
sp=np.split(arr,[4,7])
print(sp)
```

```python
arr=np.arange(24).reshape(6,4)
print(arr)
sp=np.split(arr,2,axis=1)
print(sp)
sp=np.split(arr,[3,5],axis=0)
print(sp)
```

```python
sp=np.vsplit(arr,[1,3])
print(sp)
sp=np.hsplit(arr,4)
print(sp)
```

```python
print(np.sqrt([2,5,67]))
print(np.exp([5,6,7]))
print(np.argmax([7,3,9]))# index of max
print(np.max([5,3,6]))
```

```python
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])

print(np.add(arr1,arr2))
print(np.divide(arr1,arr2))
print(arr1+arr2)
print(arr2%arr1)# remian     
print(arr2//arr1)# multiple
print(arr2**arr1)# power
```

```python
# times matrix
a=np.array([[1,2],[5,6]])
b=np.array([[4,5],[7,8]])
print(a@b)
```

```python
arr=np.arange(12).reshape(3,4)
view=arr[0:2]
print(view)
view[0,0]=999
print(view)
print(arr)# oraginal data is motified  
```

```python
arr=np.arange(12).reshape(3,4)
view=arr[0:2].copy()# copy to a brand new data
print(view)
view[0,0]=999
print(view)
print(arr)
```
