# LIST

```python
#creat a list
fam=['donna',1.45,'cici',1.53,'mica',1.64,'zhanghen',2.41]
print(fam)
```

```python
#index of list
print(fam[1:4])
```

```python
#replace the value
fam[0]='lucas'
print(fam)
```

```python
#add the value
fam.append('rui')
print(fam)
fam.insert(0,'anna')
print(fam)
```

```python
#delete a value in the list
del fam[0]
print(fam)
```

```python
#remove a value
fam.pop()
print(fam)
fam.pop(1)
print(fam)
fam.remove('cici')
print(fam)
```

```python
#looping the list
cats=['mimi','xuxu','dudu','didi']
for cat in cats:
    print(cat+' in a lovely bog~')
```

```python
#looping a range
area=[]
for value in range(1,21,2):
    area.append(value**2)
print(area)
```

# DICTIONARY

```python
alien={'color':'green','point':5}
print(alien['color'])
print(alien['point'])
```

```python
#add new element
alien['x_position']=3
alien['y_position']=42
print(alien)
```

```python
#modify new element
alien['x_position']=8765
alien['y_position']=442
print(alien)
```

```python
#remove new element
print('the alien is '+alien['color'])
alien['color']='yellow'
print('the alien is '+alien['color'])
```

```python
del alien['point']
print(alien)
```
