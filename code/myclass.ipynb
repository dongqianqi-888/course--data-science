```python
class data_master:
    '''This module in order to help you to be a data analysis master!
    Every time we have a plan for learning study data analysis, carring it out becomes a difficulty. 
    The purpose of this class is to record the time you spend on different projects and scores we obtain, 
    quantifying the abstract process of implementing the plan into a number,
    so as to better supervise our implementation and adjust the plan.'''
    def __init__(self,date):
        self.date=date
        self.hour=0
        self.grade=0
        self.sumhour=0
        self.sumgrade=0
    def check_project(self):
        print('You could learn python, r and statistic.')
    def python(self,hour,grade):
        self.hour=hour
        self.grade=grade
        self.sumhour+=hour
        self.sumgrade+=grade
        print('You spend '+str(self.hour)+' hours getting '+str(self.grade)+' grade in python!')
    def statistic(self,hour,grade):
        self.hour=hour
        self.grade=grade
        self.sumhour+=hour
        self.sumgrade+=grade
        print('You spend '+str(self.hour)+' hours getting '+str(self.grade)+' grade in statistic!')
    def r(self,hour,grade):
        self.hour=hour
        self.grade=grade
        self.sumhour+=hour
        self.sumgrade+=grade
        print('You spend '+str(self.hour)+' hours getting '+str(self.grade)+' grade in r!')
    def summary(self):
        print('\n'+self.date+':\nYou spend '+str(self.sumhour)+' hours to study, and get '+str(self.sumgrade)+' grade in total! You will surely be a good data analyst')
```
